```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas_datareader.data import DataReader
from datetime import date
```

```python
series_code = "DGS10"
data_source = "fred"
start = date(1965,1,1)

data_10_treasury = DataReader(series_code,data_source, start=start)
```

```python
data_10_treasury.info()
```

```python
data_10_treasury.head()
```

```python
series_name = "10-year Treasury"
data_10_treasury = data_10_treasury.rename(columns = {series_code:series_name})
```

```python
data_10_treasury.head()
```

```python
data_10_treasury.plot(title = series_name,figsize=(20,10)); plt.show()
```
